```python
url = "https://pt.soccerstats247.com/competi%c3%a7%c3%b5es/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
painel = soup.select("div[class='panel-default']")
paises = {}
ligas = []
for i, pais in enumerate(painel):
    link_img = pais.select('img')[0].attrs['src']
    nome_pais = pais.select('img')[0].attrs['alt']
    tr = pais.select('table tr')
    for t in tr:
        liga = {}
        l = t.select('a')[0].text
        a = t.select('a')[0].attrs['href']
        liga[l] = a
        ligas.append(liga)
    paises[nome_pais] = {'bandeira':link_img, 'ligas':ligas[:]}
    ligas.clear()
pprint(paises)
```
